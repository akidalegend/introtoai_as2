# Image Classification Using Convolutional Neural Networks

## Problem Statement
The objective of this project is to classify images of individuals into three categories: `person1`, `person2`, and `person3`. The dataset consists of images collected and organized into subdirectories based on the class labels.

## Machine Learning Approach
A Convolutional Neural Network (CNN) is employed to tackle this image classification task. CNNs are well-suited for image processing tasks due to their ability to learn spatial hierarchies of features directly from pixel data.

## Design Decisions
1. **Image Preprocessing**: Images are resized to 64x64 pixels and normalized to a [0, 1] range for consistency and faster convergence.
2. **Model Architecture**: The CNN includes convolutional layers with ReLU activation, max-pooling layers for dimensionality reduction, dropout layers to prevent overfitting, and fully connected layers for classification.
3. **Evaluation Metrics**: Accuracy, precision, recall, and confusion matrices are used to assess model performance.
4. **References**:
   - Deep Learning textbook by Ian Goodfellow et al.
   - Online tutorials on CNN implementation in TensorFlow/Keras.


# Import Required Libraries
Import necessary libraries such as NumPy, Matplotlib, TensorFlow, and OpenCV.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.utils import class_weight
from tensorflow import keras
from tensorflow.keras import layers
```

# Load and Preprocess Dataset
Load the image dataset, resize images, and normalize pixel values.

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Example of assigning labels based on subdirectories
classes = ['person1', 'person2', 'person3']
image_dir = '/Users/akramcareem/Desktop/untitled folder'

# Initialize lists to hold images and labels
images = []
labels = []

# Define the target size for the images
target_size = (128, 128)  # You can adjust this size

for idx, person in enumerate(classes):
    person_dir = os.path.join(image_dir, person)
    for filename in os.listdir(person_dir):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            img_path = os.path.join(person_dir, filename)
            img = load_img(img_path, target_size=(128, 128))
            img_array = img_to_array(img) / 255.0
            images.append(img_array)
            labels.append(idx)  # Use the index as the class label


# Convert lists to numpy arrays
images = np.array(images)
labels = np.array(labels)

# Encode labels to integers
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(labels)

# Convert labels to categorical
labels = to_categorical(labels)
```

```python
import matplotlib.pyplot as plt

# Display a sample grid of images
def display_sample_images(images, labels, class_names, num_samples=5):
    fig, axes = plt.subplots(1, num_samples, figsize=(15, 5))
    for i in range(num_samples):
        axes[i].imshow(images[i])
        axes[i].set_title(f"Label: {class_names[np.argmax(labels[i])]}")
        axes[i].axis("off")
    plt.tight_layout()
    plt.show()

# Example usage (ensure you have class names defined)
class_names = ['person1', 'person2', 'person3']
display_sample_images(images, labels, class_names)

# Class distribution
import seaborn as sns

class_counts = [np.sum(labels[:, i]) for i in range(labels.shape[1])]
sns.barplot(x=class_names, y=class_counts)
plt.title("Class Distribution")
plt.ylabel("Number of Samples")
plt.xlabel("Class")
plt.show()
```

# Split Dataset into Training and Testing Sets
Use train_test_split to divide the dataset into training and testing sets.

```python
# Split Dataset into Training and Testing Sets
num_classes = 3 # Number of classes

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)

# Ensure y_train and y_test are 1D arrays of class indices
y_train = np.array(y_train).flatten()
y_test = np.array(y_test).flatten()

# Convert labels to one-hot encoding
y_train_cat = to_categorical(y_train, num_classes=num_classes)
y_test_cat = to_categorical(y_test, num_classes=num_classes)

```

# Build the CNN Model
Define a Convolutional Neural Network (CNN) model using TensorFlow's Sequential API.

```python
# # Update the model's final layer
# model = Sequential([
#     Conv2D(32, (6, 6), activation='relu', input_shape=(128, 128, 3)),
#     MaxPooling2D((2, 2)),
#     Conv2D(128, (3, 3), activation='relu'),
#     MaxPooling2D((2, 2)),
#     Conv2D(128, (3, 3), activation='relu'),
#     MaxPooling2D((2, 2)),
#     Flatten(),
#     Dense(128, activation='relu'),
#     Dropout(0.5),
#     Dense(num_classes, activation='softmax')  # Adjusted for 3 classes
# ])

inputs = keras.Input(shape=(180, 180, 3))
x = layers.Rescaling(1./255)(inputs)
x = layers.Conv2D(filters=32, kernel_size=3, activation="relu")(x)
x = layers.MaxPooling2D(pool_size=2)(x)
x = layers.Conv2D(filters=64, kernel_size=3, activation="relu")(x)
x = layers.MaxPooling2D(pool_size=2)(x)
x = layers.Conv2D(filters=128, kernel_size=3, activation="relu")(x)
x = layers.MaxPooling2D(pool_size=2)(x)
x = layers.Conv2D(filters=256, kernel_size=3, activation="relu")(x)
x = layers.MaxPooling2D(pool_size=2)(x)
x = layers.Conv2D(filters=256, kernel_size=3, activation="relu")(x)
x = layers.Flatten()(x)
x = layers.Dropout(0.5)(x)
outputs = layers.Dense(num_classes, activation="softmax")(x)  # Adjusted for multi-class classification
model = keras.Model(inputs=inputs, outputs=outputs)

# Compile the model
#model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.compile(loss="categorical_crossentropy",
              optimizer="rmsprop",
              metrics=["accuracy"])


# Display the model architecture
model.summary()

# Print the number of files processed in each class
print(f"Number of files processed in each class: {class_counts}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Define the model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D((2, 2)),
    Dropout(0.25),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Dropout(0.25),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(class_names), activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    images, labels, epochs=10, validation_split=0.2, batch_size=32, verbose=2
)

# Plot accuracy and loss
def plot_metrics(history):
    plt.figure(figsize=(12, 5))
    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy')
    plt.legend()

    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss')
    plt.legend()

    plt.show()

plot_metrics(history)

# Evaluate the model
y_pred = model.predict(images)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(labels, axis=1)

# Classification Report
print("Classification Report:")
print(classification_report(y_true_classes, y_pred_classes, target_names=class_names))
print(num_classes)
```

# Data Augmentation
Use ImageDataGenerator to apply data augmentation techniques to the training dataset.

```python
# Data Augmentation

# Initialize the ImageDataGenerator with data augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)


# Use flow with X_train and y_train_cat
train_generator = datagen.flow(X_train, y_train, batch_size=32)

print('Model output shape:', model.output_shape)       # Expected: (None, 3)
print('y_train_cat shape:', y_train_cat.shape)         # Expected: (num_samples, 3)
print('y_test_cat shape:', y_test_cat.shape)

# Fit the ImageDataGenerator to the training data
datagen.fit(X_train)

# Example of how to use the ImageDataGenerator to augment a batch of images
# This is just for demonstration; in practice, you would use this in model training
for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=32):
    # Plot the first 9 images in the batch
    for i in range(0, 9):
        plt.subplot(330 + 1 + i)
        plt.imshow(X_batch[i])
    plt.show()
    break  # Stop after the first batch for demonstration purposes

print(num_classes)
```

# Compile the Model
Compile the CNN model with appropriate loss function, optimizer, and evaluation metrics.

```python
# Compile the Model

# Compile the CNN model with appropriate loss function, optimizer, and evaluation metrics
model.compile(
    optimizer='adam',  # Use Adam optimizer
    loss='categorical_crossentropy',  # Use categorical crossentropy loss for multi-class classification
    metrics=['accuracy']  # Evaluate the model using accuracy metric
)

# Display the model architecture
model.summary()


```

# Train the Model
Train the CNN model using the training dataset and validate it using the testing dataset.

```python
# Train the Model

# Define early stopping callback to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model using the training dataset and validate it using the testing dataset
history = model.fit(
    datagen.flow(X_train, y_train, batch_size=32),
    epochs=100,
    validation_data=(X_test, y_test),
    callbacks=[early_stopping]
)

```

# Evaluate the Model
Evaluate the trained model on the testing dataset and display the results.

```python
# Evaluate the Model

# Evaluate the trained model on the testing dataset
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=2)

print(f'Test Loss: {test_loss}')
print(f'Test Accuracy: {test_accuracy}')
```
