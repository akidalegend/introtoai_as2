# Image Classification Using Convolutional Neural Networks

## Problem Statement
The objective of this project is to classify images of individuals into three categories: `person1`, `person2`, and `person3`. The dataset consists of images collected and organized into subdirectories based on the class labels.

## Machine Learning Approach
A Convolutional Neural Network (Sequential and Functional) and a Siamnese Network is employed to tackle this image classification task. CNNs are well-suited for image processing tasks due to their ability to learn spatial hierarchies of features directly from pixel data.

## Design Decisions
1. **Image Preprocessing**: Images are resized to 64x64 pixels and normalized to a [0, 1] range for consistency and faster convergence.
2. **Model Architecture**: The CNN includes convolutional layers with ReLU activation, max-pooling layers for dimensionality reduction, dropout layers to prevent overfitting, and fully connected layers for classification.
3. **Evaluation Metrics**: Accuracy, precision, recall, and confusion matrices are used to assess model performance.
4. **References**:
   - Deep Learning textbook by Ian Goodfellow et al.
   - Online tutorials on CNN implementation in TensorFlow/Keras.


# Import Required Libraries
Import necessary libraries such as NumPy, Matplotlib, TensorFlow, and OpenCV.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.utils import class_weight
from tensorflow import keras
from tensorflow.keras import layers
```

# Load and Preprocess Dataset
Load the image dataset, resize images, and normalize pixel values.

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Example of assigning labels based on subdirectories
classes = ['person1', 'person2', 'person3']
image_dir = '/Users/akramcareem/Desktop/untitled folder'

# Initialize lists to hold images and labels
images = []
labels = []

# Define the target size for the images
target_size = (256, 256)  # Size can be adjusted

for idx, person in enumerate(classes):
    person_dir = os.path.join(image_dir, person)
    for filename in os.listdir(person_dir):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            img_path = os.path.join(person_dir, filename)
            img = load_img(img_path, target_size=target_size)
            img_array = img_to_array(img) / 255.0
            images.append(img_array)
            labels.append(idx)  # Using the index as the class label


# Convert lists to numpy arrays
images = np.array(images)
labels = np.array(labels)

# Encode labels to integers
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(labels)

# Convert labels to categorical
labels = to_categorical(labels)
```

```python
# Class distribution
import seaborn as sns
class_counts = [np.sum(labels[:, i]) for i in range(labels.shape[1])]
class_names = ['person1', 'person2', 'person3']
sns.barplot(x=class_names, y=class_counts)
plt.title("Class Distribution")
plt.ylabel("Number of Samples")
plt.xlabel("Class")
plt.show()
```

# Split Dataset into Training and Testing Sets
Use train_test_split to divide the dataset into training and testing sets.

```python
# Split Dataset into Training and Testing Sets
num_classes = 3 # Number of classes

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)

# Ensure y_train and y_test are 1D arrays of class indices
y_train = np.array(y_train).flatten()
y_test = np.array(y_test).flatten()

# Convert labels to one-hot encoding
y_train_cat = to_categorical(y_train, num_classes=num_classes)
y_test_cat = to_categorical(y_test, num_classes=num_classes)

```

# 1. Sequential Convolutional Neural Network (CNN)
Define a Convolutional Neural Network (CNN) model using TensorFlow's Sequential API.

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Define the model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3)),
    MaxPooling2D((2, 2)),
    Dropout(0.25),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Dropout(0.5),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(class_names), activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()

# Train the model
history = model.fit(
    images, labels, epochs=50, validation_split=0.2, batch_size=32, verbose=2
)

# Plot accuracy and loss
def plot_metrics(history):
    plt.figure(figsize=(12, 5))
    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy')
    plt.legend()

    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss')
    plt.legend()

    plt.show()

plot_metrics(history)
```

```python
# Evaluate the model
y_pred = model.predict(images)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(labels, axis=1)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=2)
print(f'Test Accuracy: {test_accuracy:.4f}')

# Predict on the test set
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# Classification report and confusion matrix
print(classification_report(y_true_classes, y_pred_classes, target_names=class_names, labels=[0, 1, 2]))

conf_matrix = confusion_matrix(y_true_classes, y_pred_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

# 2. Functional Model

```python
inputs = keras.Input(shape=(256, 256, 3))
x = layers.Rescaling(1./255)(inputs) # Rescale the pixel values to the [0, 1] range
x = layers.Conv2D(filters=32, kernel_size=3, activation="relu")(x) # New convolutional layer
x = layers.MaxPooling2D(pool_size=2)(x) # New max pooling layer
x = layers.Conv2D(filters=64, kernel_size=3, activation="relu")(x) # New convolutional layer
x = layers.MaxPooling2D(pool_size=2)(x) # New max pooling layer
x = layers.Conv2D(filters=128, kernel_size=3, activation="relu")(x) # New convolutional layer
x = layers.MaxPooling2D(pool_size=2)(x) # New max pooling layer
x = layers.Conv2D(filters=256, kernel_size=3, activation="relu")(x) # New convolutional layer
x = layers.MaxPooling2D(pool_size=2)(x) # New max pooling layer
x = layers.Conv2D(filters=256, kernel_size=3, activation="relu")(x) # New convolutional layer
x = layers.Flatten()(x) # Flatten the output of the last convolutional layer before the dense layers
x = layers.Dropout(0.5)(x) # Add dropout for regularization
outputs = layers.Dense(num_classes, activation="softmax")(x)  # Adjusted for multi-class classification
model = keras.Model(inputs=inputs, outputs=outputs)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    images, labels, epochs=40, validation_split=0.2, batch_size=42, verbose=2
)

# Plotting the curve of loss and accuracy during training
def plot_training_history(history):
    plt.figure(figsize=(12, 5))

    # Plot training & validation accuracy values
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    # Plot training & validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.show()

plot_training_history(history)

# Display the model architecture
model.summary()


```

```python
# Evaluate the model
y_pred = model.predict(images)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(labels, axis=1)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=2)
print(f'Test Accuracy: {test_accuracy:.4f}')

# Predict on the test set
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# Classification report and confusion matrix
print(classification_report(y_true_classes, y_pred_classes, target_names=class_names, labels=[0, 1, 2]))

conf_matrix = confusion_matrix(y_true_classes, y_pred_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

# 3. Siamese Networks 

```python
from tensorflow.keras.models import Model
import tensorflow.keras.backend as K
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Lambda
from tensorflow.keras.preprocessing.image import img_to_array, load_img
import numpy as np
import cv2

def siamese_network(input_shape):
    input = Input(shape=input_shape)
    x = Conv2D(128, (10, 10), activation='relu')(input)
    x = MaxPooling2D()(x)
    x = Conv2D(128, (7, 7), activation='relu')(x)
    x = MaxPooling2D()(x)
    x = Conv2D(128, (4, 4), activation='relu')(x)
    x = MaxPooling2D()(x)
    x = Conv2D(256, (4, 4), activation='relu')(x)
    x = Flatten()(x)
    x = Dense(4096, activation='sigmoid')(x)
    return Model(input, x)

def euclidean_distance(vectors):
    (featsA, featsB) = vectors
    sum_squared = K.sum(K.square(featsA - featsB), axis=1, keepdims=True)
    return K.sqrt(K.maximum(sum_squared, K.epsilon()))

input_shape = (105, 105, 1)
base_network = siamese_network(input_shape)

inputA = Input(shape=input_shape)
inputB = Input(shape=input_shape)

featsA = base_network(inputA)
featsB = base_network(inputB)

distance = Lambda(euclidean_distance)([featsA, featsB])
model = Model(inputs=[inputA, inputB], outputs=distance)

model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Resize and convert images to grayscale
def preprocess_images(images):
    processed_images = []
    for img in images:
        # Resize and convert to grayscale using OpenCV
        img = cv2.resize(img, (105, 105))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = img / 255.0  # Normalize the image
        processed_images.append(img)
    return np.array(processed_images).reshape(-1, 105, 105, 1)  # Ensure the shape is (num_samples, 105, 105, 1)

# Preprocess the training and testing images
X_train_processed = preprocess_images(X_train)
X_test_processed = preprocess_images(X_test)

# Create pairs of images and corresponding labels
def create_pairs(images, labels):
    pairs = []
    pair_labels = []
    num_classes = len(np.unique(labels))
    for idx1 in range(len(images)):
        for idx2 in range(idx1 + 1, len(images)):
            pairs.append([images[idx1], images[idx2]])
            if labels[idx1] == labels[idx2]:
                pair_labels.append(1)
            else:
                pair_labels.append(0)
    return np.array(pairs), np.array(pair_labels)

# Create pairs for training and testing
X_train_pairs, y_train_pairs = create_pairs(X_train_processed, np.argmax(y_train, axis=1))
X_test_pairs, y_test_pairs = create_pairs(X_test_processed, np.argmax(y_test, axis=1))

# Train the model
history = model.fit(
    [X_train_pairs[:, 0], X_train_pairs[:, 1]], y_train_pairs, 
    validation_data=([X_test_pairs[:, 0], X_test_pairs[:, 1]], y_test_pairs),
    batch_size=32, epochs=20, verbose=2
)

model.summary()
# Plotting the training and validation loss and accuracy
def plot_training_history(history):
    plt.figure(figsize=(12, 5))

    # Plot training & validation accuracy values
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

plot_training_history(history)
```

```python
# Evaluate the model
y_pred = model.predict(images)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(labels, axis=1)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=2)
print(f'Test Accuracy: {test_accuracy:.4f}')

# Predict on the test set
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# Classification report and confusion matrix
print(classification_report(y_true_classes, y_pred_classes, target_names=class_names, labels=[0, 1, 2]))

conf_matrix = confusion_matrix(y_true_classes, y_pred_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

# Performance Metrics Discussion


#### Choice of Performance Metrics

**Accuracy:**
Accuracy is one of the primary metrics used to evaluate the performance of our facial recognition models. It measures the proportion of correctly classified instances (both true positives and true negatives) out of the total instances. While accuracy provides a general sense of the model's performance, it may not fully capture the nuances of the application, especially in cases of class imbalance.

**Precision and Recall:**
Precision and recall are crucial metrics for evaluating the performance of facial recognition systems, particularly in scenarios where false positives and false negatives have different implications. Precision measures the proportion of true positive predictions out of all positive predictions made by the model, indicating how many of the predicted positive instances are actually correct. Recall, on the other hand, measures the proportion of true positive predictions out of all actual positive instances, indicating how well the model captures all relevant instances.

**F1-Score:**
The F1-score is the harmonic mean of precision and recall, providing a single metric that balances both. It is particularly useful when there is a need to balance the trade-off between precision and recall. A high F1-score indicates that the model performs well in both identifying relevant instances and minimizing false positives.

**Confusion Matrix:**
The confusion matrix provides a detailed breakdown of the model's performance by showing the counts of true positives, true negatives, false positives, and false negatives. It helps in understanding the types of errors the model is making and identifying any potential biases or weaknesses in the model's predictions.


#### Relevance to the Application Scenario

**Security and Surveillance:**
In security and surveillance applications, the cost of false positives (incorrectly identifying an innocent person as a suspect) and false negatives (failing to identify a suspect) can be significant. Precision and recall are critical metrics in this context. High precision ensures that the system does not falsely accuse individuals, while high recall ensures that the system effectively identifies all potential suspects.

**Access Control:**
For access control systems, such as those used in secure facilities or personal devices, both precision and recall are important. High precision ensures that unauthorized individuals are not granted access, while high recall ensures that authorized individuals are not denied access. The F1-score provides a balanced measure of the system's performance in this scenario.

**Social Media and Personalization:**
In social media applications, where facial recognition is used for tagging and organizing photos, user experience is paramount. High accuracy and precision are important to ensure that users are correctly identified and tagged in photos. False positives can lead to incorrect tagging, while false negatives can result in missed tagging opportunities.

**Ethical Considerations:**
The choice of performance metrics also has ethical implications. Ensuring high precision and recall across different demographic groups is essential to avoid biases and ensure fairness. The confusion matrix can help identify any disparities in the model's performance across different groups, allowing for targeted improvements.

#### Conclusion

The choice of performance metrics is critical to evaluating and improving the effectiveness of facial recognition models. Accuracy, precision, recall, F1-score, confusion matrix, and ROC-AUC are all relevant metrics that provide different insights into the model's performance. By carefully selecting and analyzing these metrics, we can ensure that our facial recognition system meets the specific requirements of the application scenario and addresses any ethical considerations related to fairness and bias.

# Bias and Ethical Challenges Analysis

#### Dataset Bias

**Representation Bias:**
The Labeled Faces in this dataset, like many other facial recognition datasets, may suffer from representation bias. This occurs when certain demographic groups (e.g., based on race, gender, age) are underrepresented or overrepresented in the dataset. For instance, if the dataset predominantly contains images of individuals from a specific ethnic background, the model trained on this dataset may perform poorly on faces from other ethnic backgrounds. This can lead to biased outcomes and reduced accuracy for underrepresented groups.

**Collection Bias:**
The images in this may have been collected from the web, which may introduce collection bias. The sources of these images might favor certain demographics or contexts (e.g., celebrities, public figures), leading to a dataset that does not accurately reflect the diversity of the general population. This can result in a model that is less effective in real-world applications where the diversity of faces is much broader.

#### Model Design Bias

**Algorithmic Bias:**
The design of the deep learning models, including the Sequential, Functional, and Siamese networks, can introduce algorithmic bias. For example, if the models are not adequately regularized or if the training process does not account for class imbalances, the models may become biased towards the majority class. This can lead to unfair treatment of minority classes and reduced overall fairness of the facial recognition system.

**Evaluation Bias:**
The metrics used to evaluate the models, such as accuracy, may not fully capture the fairness and ethical implications of the model's performance. For instance, a model with high overall accuracy might still exhibit significant disparities in performance across different demographic groups. It is essential to use additional metrics, such as precision, recall, and fairness metrics, to evaluate the model's performance more comprehensively.

#### Ethical Challenges

**Privacy Concerns:**
Facial recognition technology raises significant privacy concerns. The use of facial recognition systems in public and private spaces can lead to unauthorized surveillance and tracking of individuals without their consent. It is crucial to implement strict privacy policies and obtain explicit consent from individuals before collecting and using their facial data.

**Misuse and Abuse:**
There is a risk of misuse and abuse of facial recognition technology by governments, organizations, and individuals. For example, facial recognition systems can be used for mass surveillance, racial profiling, and other forms of discrimination. It is essential to establish clear ethical guidelines and legal frameworks to prevent the misuse of this technology and protect individuals' rights.

**Transparency and Accountability:**
Ensuring transparency and accountability in the development and deployment of facial recognition systems is critical. Developers and organizations must be transparent about the data sources, model design, and evaluation processes. Additionally, there should be mechanisms in place to hold developers and users accountable for any biased or unethical outcomes resulting from the use of facial recognition technology.

#### Mitigation Strategies

**Diverse and Representative Datasets:**
To mitigate representation bias, it is essential to use diverse and representative datasets that include a wide range of demographic groups. Efforts should be made to collect and curate datasets that reflect the diversity of the population.

**Fairness-Aware Algorithms:**
Incorporating fairness-aware algorithms and techniques, such as re-sampling, re-weighting, and adversarial debiasing, can help reduce algorithmic bias. These techniques aim to ensure that the model's performance is fair and unbiased across different demographic groups.

**Ethical Guidelines and Legal Frameworks:**
Establishing and adhering to ethical guidelines and legal frameworks is crucial to address the ethical challenges associated with facial recognition technology. These guidelines should emphasize the importance of privacy, consent, transparency, and accountability.

**Continuous Monitoring and Evaluation:**
Continuous monitoring and evaluation of the model's performance and ethical implications are necessary to identify and address any emerging biases or ethical issues. Regular audits and assessments can help ensure that the facial recognition system remains fair, unbiased, and ethically sound.

By critically analyzing the bias and ethical challenges related to the dataset and model design, we can develop more fair, transparent, and accountable facial recognition systems that respect individuals' rights and promote ethical use of technology.
